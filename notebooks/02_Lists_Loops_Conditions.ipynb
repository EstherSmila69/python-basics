## 1. Lists in Python
### ðŸ”¶ Task 1.1 â€“ Create a weekly SST list

```python
# TODO: Create your SST list and print first, last, and length.
sst = [28.1, 28.3, 28.3, 27.9, 28.5, 28.7, 28.3]

print("First value:", sst[0])
print("Last value:", sst[-1])
print("Length of list:", len(sst))
```

## 2. Loops (for-loops)
### ðŸ”¶ Task 2.1 â€“ Print each value

```python
# TODO: Print each SST value with day number.
for i in range(len(sst)):
    temp = sst[i]
    # print Day X: value
    print(f"Day {i+1}: {temp}")

```

## 3. Computing Mean using Loop
### ðŸ”¶ Task 3.1 â€“ Compute mean SST manually

```python
total = 0

for temp in sst:
    total += temp

mean_sst = total / len(sst)
print("Mean SST (loop):", mean_sst)
print("Mean SST (built-in):", sum(sst)/len(sst))
```

## 4. Conditions: HOT or NORMAL days
### ðŸ”¶ Task 4.1 â€“ Classify SST days

```python
threshold = 28.5

for i in range(len(sst)):
    temp = sst[i]
    day = i + 1
    # TODO: write if condition here
if temp > threshold:
        print(f"Day {day}: SST = {temp} (Hot days)")
else:
        print(f"Day {day}: SST = {temp} (Normal days)")
```

## 5. Filtering SST values
### ðŸ”¶ Task 5.1 â€“ Create HOT and NORMAL SST lists

```python
hot_sst = []
normal_sst = []

for temp in sst:
    if temp > threshold:
        hot_sst.append(temp)
    else:
        normal_sst.append(temp)


print("Hot SST values:", hot_sst)
print("Normal SST values:", normal_sst)
```

## 6. Mini Task â€“ Rainfall classification
### ðŸ”¶ Task 6.1 â€“ Classify rainfall intensity

```python
rain = [0.0, 12.5, 34.0, 2.0, 0.0, 55.1, 8.4]

no_rain = 0
light = 0
moderate = 0
heavy = 0

for r in rain:
    if r == 0.0:
        no_rain += 1
    elif 0 < r <= 10:
        light += 1
    elif 10 < r <= 50:
        moderate += 1
    else:  # r > 50
        heavy += 1

print("No rain days:", no_rain)
print("Light rain days:", light)
print("Moderate rain days:", moderate)
print("Heavy rain days:", heavy)
```
